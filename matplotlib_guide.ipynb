#### Changing the font properties of texts associated with the chart

```python
fig = plt.figure(1, figsize=(6,6))

ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

plt.title('Raining Hogs and Dogs')
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
fracs = [15,30,45, 10]

patches, texts, autotexts = ax.pie(fracs, labels=labels, autopct='%1.1f%%')

proptease = fm.FontProperties()
proptease.set_size('x-large') # Selecting a font size

plt.setp(autotexts, fontproperties=proptease) # Setting font size to the percentage values of the slices
plt.setp(texts, fontproperties=proptease) # Setting font size to the labels of the slices

plt.show()
```

### Stack Plot

```python
plt.stackplot(experience, data_scientists_salary, colors = 'g')

plt.title('Salary of Data Scientists by their experience')
plt.xlabel('Experience')
plt.ylabel('Salary')

plt.show()
```

In Stack plot, the values for each plot get added on top of each other. So the plots are not overlapping each other, as we saw in the case of multiple bar plots.

#### Example 1

```python
days = [1,2,3,4,5,6]

sleep = [6,7,5,8,6,7]
drinking_water =   [2,2,1,2,1,1]
work =  [5,7,10,8,6,9]
exercise=  [3,3,0,1,3,2]
```

```python
# To be able to add legends, we need to provide empty plots, with labels and corresponding colors codes
plt.plot([],[],color='green', label='sleep', linewidth=3)
plt.plot([],[],color='blue', label='drinking_water', linewidth=3)
plt.plot([],[],color='red', label='work', linewidth=3)
plt.plot([],[],color='orange', label='play', linewidth=3)

# Visible Stack Plot
plt.stackplot(days, sleep, drinking_water, work, exercise, colors=['g', 'b', 'r', 'orange'])

plt.xlabel('Days')
plt.ylabel('Activities')
plt.title('6 Day Routine stack plot')
plt.legend(loc='lower right')
plt.grid(True)

plt.show()
```

#### Example 2

```python
stock1= [5,3,3,6,1,8,2,7,9]
stock2= [2,4,1,3,5,0,3,1,0]
stock3= [2,2,5,0,3,1,4,1,0]
# Effective sum of all the 3 stocks
# sum = [9,9,9,9,9,9,9,9,9]

# Thus we can expect a horizontal line at the top of the stack plot

days =[1,2,3,4,5,6,7,8,9]
```

```python
stocks = ['stock 1', 'stock 2', 'stock 3'] # Better Approach to add labels for legend
colors = ["#F9CDAD", "#FC9D9A", "#83AF9B"]

plt.title('Stock Rates')

plt.stackplot(days, stock1, stock2, stock3, labels=stocks, colors=colors)

plt.show()
```

### Contour (Level) Plot

Contour plots (sometimes called Level Plots) are a way to show a three-dimensional surface on a two-dimensional plane. It graphs two predictor variables X Y on the y-axis and a response variable Z as contours. These contours are sometimes called the z-slices or the iso-response values.

```python
x = [0,3,6,9,13,15,19,23,26,29,33,35,39,41,47,56] 
y = [5,8,13,16,17,20,25,26,30,33,37,39,41,44,48,59]
```

```python
[X, Y] = np.meshgrid(x, y) # Creating 2-D grid of features

Z = np.sqrt(X**2 + Y**2)

plt.contour(X, Y, Z)

plt.title('Contour Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values')

plt.show()
```

#### Filling inside the plot

```python
[X, Y] = np.meshgrid(x, y) # Creating 2-D grid of features

Z = np.sqrt(X**2 + Y**2)

plt.contourf(X, Y, Z)

plt.title('Contour Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values')

plt.show()
```

### Plotting Time Series

```python
dates = [
    datetime(2021, 3, 10),
    datetime(2021, 3, 13),
    datetime(2021, 3, 14),
    datetime(2021, 3, 15),
    datetime(2021, 3, 16),
    datetime(2021, 3, 17),
    datetime(2021, 3, 18),
    datetime(2021, 3, 19)
]

values = [0,3,4,7,5,3,5,6]
```

```python
plt.title('Time Series')

plt.plot_date(dates, values)
plt.xticks(rotation='vertical')
plt.grid(True)
plt.show()
```

```python
plt.title('Time Series')

# ms - marker size
# mfc - marker face color
# c - color of the line
plt.plot_date(dates, values, linestyle='solid', marker='o', ms=10, mfc='r', c='b', linewidth=3)

plt.xticks(rotation='vertical')
plt.xlabel('Dates')
plt.ylabel('Values')

plt.grid(True)
plt.show()
```

### Heatmap

A heat map is a data visualization technique that shows magnitude of a phenomenon as color in two dimensions. The variation in color may be by hue or intensity, giving obvious visual cues to the reader about how the phenomenon is clustered or varies over space. It's generally used to understand correlations between variables.

```python
data = np.random.random(( 6 , 6 )) # 2-D Data
```

```python
plt.imshow(data, cmap='autumn')
plt.title('2-D Heat Map')
plt.show()
```

```python
plt.pcolormesh(data, cmap='coolwarm')
plt.title('2-D Heat Map')
plt.show()
```

#### Real world example

```python
import seaborn as sns
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights = flights.pivot('month', 'year', 'passengers')
flights
```

```python
plt.pcolormesh(flights, cmap='YlGn', edgecolor='w')

cbar = plt.colorbar()

plt.title('Flights Heat Map')
plt.xlabel('Year')
plt.ylabel('Month')

# ticks: set the position of the axis values
# labels: set the label for each tick
plt.xticks(ticks=np.arange(len(flights.columns)) + 0.5, labels=list(flights.columns), rotation = 90)
plt.yticks(ticks=np.arange(len(flights.index)) + 0.5, labels=list(flights.index))
plt.show()
```

```python
plt.imshow(flights, cmap='YlGn')

cbar = plt.colorbar()

plt.title('Flights Heat Map')
plt.xlabel('Year')
plt.ylabel('Month')

plt.xticks(ticks=np.arange(len(flights.columns)) + 0.5, labels=list(flights.columns), rotation = 90)
plt.yticks(ticks=np.arange(len(flights.index)) + 0.5, labels=list(flights.index))
plt.grid(True, which='both')

plt.show()
```
